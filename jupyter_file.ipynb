```python
import os
import pandas as pd
from pathlib import Path
```

```python
# create the paths using Path

bit_data = Path('Resources/bitstamp.csv')
coin_data = Path('Resources/coinbase.csv')
bit_data
coin_data
```

```python
# create the bitstamp data frame 

bitstamp = pd.read_csv(bit_data, index_col="Timestamp", parse_dates=True, infer_datetime_format=True)
bitstamp.head()
```

```python
# create the coinbase data frame 

coinbase = pd.read_csv(coin_data, index_col="Timestamp", parse_dates=True, infer_datetime_format=True)
coinbase.head()
```

```python
# drop NAs
# review duplicates
bitstamp[bitstamp.duplicated(keep=False)]
# no duplicates found         

bitstamp.dropna(inplace=True)
bitstamp.drop_duplicates(inplace=True) # should not actually do anything
```

```python
bitstamp['Close'] = bitstamp['Close'].str.replace('$','',regex=False).astype(float)
bitstamp.head()
coinbase['Close'] = coinbase['Close'].str.replace('$','',regex=False).astype(float)
coinbase.head()
```
